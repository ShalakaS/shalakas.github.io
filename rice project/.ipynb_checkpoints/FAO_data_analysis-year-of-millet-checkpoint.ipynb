## Wheat

```python
wheat_df = df[df.Item == "Wheat", ]
wheat_df = wheat_df.reset_index(drop="index")
wheat_df
```

## Corn

```python
corn_df = df[df.Item == "Maize (corn)"]
corn_df = corn_df.reset_index(drop="index")
corn_df
```

## Combined

```python
all_grains = pd.concat([wheat_df, rice_df, millet_df, corn_df], axis=0, ignore_index=True)
all_grains
```
